```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Load the dataset, assign column names manually
column_names = ['ID', 'Topic', 'Sentiment', 'Text']
df = pd.read_csv("C:/Users/Admin/Downloads/twitter_training.csv", header=None, names=column_names)

# Check the first few rows of the data to ensure it looks correct
print(df.head())
```

```python
print(df.info())  # Check column names and data types
```

```python

# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())

```

```python
# Check data types of columns
print("\nData types of each column:")
print(df.dtypes)
```

```python
# Fill missing values in the 'Text' column with a placeholder string
df['Text'].fillna('No text available', inplace=True)

# Check the result
print(df.isnull().sum())  # To ensure there are no more missing values
print(df.head())  # To see the changes made
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
# Convert 'Topic' and 'Sentiment' columns to categorical data type
df['Topic'] = df['Topic'].astype('category')
df['Sentiment'] = df['Sentiment'].astype('category')

# Check the updated data types
print(df.dtypes)
```

```python
# Perform sentiment analysis (though sentiment is already labeled in the 'Sentiment' column)
# If you'd like to double-check sentiment with VADER, follow this procedure:
sia = SentimentIntensityAnalyzer()
```

```python
# Function to calculate sentiment polarity (if you want to compare with the existing sentiment column)
def get_sentiment(text):
    score = sia.polarity_scores(text)
    if score['compound'] >= 0.05:
        return 'positive'
    elif score['compound'] <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis to the 'Text' column
df['predicted_sentiment'] = df['Text'].apply(get_sentiment)

# Compare original 'Sentiment' column with predicted 'predicted_sentiment' to check for accuracy
print(df[['Sentiment', 'predicted_sentiment']].head())
```

```python
# Visualize the sentiment distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Sentiment', data=df, palette='viridis')
plt.title('Sentiment Distribution in Twitter Data')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```

Interpretation:
The graph illustrates the overall distribution of sentiments (Irrelevant, Negative, Neutral, and Positive) in the Twitter dataset labeled as "twitter_training." Each bar represents the total count of tweets categorized into the respective sentiment. Key observations include:

Negative Sentiment Prevalence:

Negative sentiments have the highest count, indicating that a significant portion of tweets expresses dissatisfaction, criticism, or negative opinions about the analyzed topics or brands.
Positive Sentiment:

Positive sentiments are the second-most prevalent, suggesting a substantial amount of favorable or appreciative opinions expressed on Twitter.
Neutral Sentiment:

Neutral tweets form a significant proportion, reflecting content that does not explicitly express a strong opinion (either positive or negative). These could be factual statements or informational tweets.
Irrelevant Sentiment:

The "Irrelevant" category has the lowest count. This likely includes tweets that do not directly relate to the analyzed topics or sentiments, highlighting the need for accurate filtering during preprocessing.

Conclusion:
The data suggests that social media users tend to express their dissatisfaction or concerns more frequently than positive sentiments. However, the nearly comparable counts of positive and neutral tweets indicate a mix of reactions across topics. This highlights the importance of addressing negative feedback while leveraging positive sentiments to improve public perception.
